```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df=pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/EPIDEMIC_DATA.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
features = df[['POPULATION DENSITY','INTERNATIONAL_TRAVEL','HEALTH_EXPENDITURE']]
target = df['OUTBREAK']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features,target, test_size=0.25, random_state=20)
```

```python
model = RandomForestClassifier(n_estimators=150, random_state=20)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
classification = classification_report(y_test, y_pred)
```

```python
print(f'Accuracy: {accuracy}')
print(f'Classification Report:')
print(classification)
```
